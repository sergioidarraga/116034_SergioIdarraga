```python
#3. LEER UN NUMERO DE TRES DIGITOS Y DETERMINAR SI CONTIENE EL DIGITO 8

entero=int(input('Ingrese un numero de 3 digitos: '))

numabs=abs(entero)
num=str(numabs)
digitos=len(num)

if digitos==3:
    for i in range(0,digitos):
        if num[i]=='8':
            print('Si contiene el digito 8')
            break
        if i==digitos-1 and num[digitos-1]!='8':
            print('No contiene el numero 8')
else:
    print('El numero no es de 3 digitos')
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 18 08:46:35 2019

@author: SergioIdarraga
"""

#6. determinar si la suma de los digitos es un fibonacci
#suma de los digitos

num=input('Ingrese un numero entero:')
suma=0
for i in range(len(num)):
    suma=suma+int(num[i])
    
#fibonacci
    
cont=0
b=1
fibonacci1=[]
fibonacci2=[]
fibonacci=[]
while b<int(num) and cont<int(num):
    cont+=b
    b+=cont
    fibonacci1.append(b)
    fibonacci2.append(cont)
    if b<=int(num) or cont<=int(num):
        fibonacci=sorted(fibonacci1+fibonacci2)
if suma in fibonacci:
    print(f'La suma {suma} pertenece a la secuencia Fibonacci')
else:
    print(f'La suma {suma} no pertenece a la secuencia Fibonacci')
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 18 08:54:29 2019

@author: SergioIdarraga
"""

#9. Ingresar dos circulos y un punto con coordenadas y redios y determinar si:
   #el punto esta dentro del circulo 1, dentro del circulo 2, dentro de los dos
   #circulos o en ningun circulo
   
x1=float(input('Ingrese la coordenada X del circulo 1: '))
y1=float(input('Ingrese la coordenada Y del circulo 1: '))
r1=float(input('Ingrese el radio del circulo 1: '))
x2=float(input('Ingrese la coordenada X del circulo 2: '))
y2=float(input('Ingrese la coordenada Y del circulo 2: '))
r2=float(input('Ingrese el radio del circulo 2: '))
a=float(input('Ingrese la coordenada en X del punto: '))
b=float(input('Ingrese la coordenada en Y del punto: '))
print('RESPUESTA')
if r1==0:
    print('El radio 1 debe ser mayor a 0')
if r2==0:
    print('El radio 2 debe ser mayor a 0')
if r1!=0 and r2!=0:
    dist1=((a-x1)**2+(b-y1)**2)**0.5 #Distancia del punto al centro 1
    dist2=((a-x2)**2+(b-y2)**2)**0.5 #Distancia del punto al centro 2
    if dist1<=r1 and dist2<=r2:
        print('El punto esta dentro de ambos circulos')
    if dist1<=r1 and dist2>r2:
        print('El punto esta dentro del circulo 1')
    if dist1>r1 and dist2<=r2:    
        print('El punto esta dentro del circulo 2')
    if dist1>r1 and dist2>r2:
        print('El punto no esta contenido en ningun circulo')
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Jun  7 17:40:26 2019

@author: SergioIdarraga
"""

# 12 Algoritmo para comprar dos palabras e informar cuál de ellas se encuentra
#primero en el diccionario

palabra1=str(input('Inserte palabra 1: '))
palabra2=str(input('Inserte palabra 2: '))

palabra1=palabra1.lower()
palabra2=palabra2.lower()

#%% Para definir rango

if len(palabra1) < len(palabra2):
    rango=len(palabra2)
    
elif len(palabra1) > len(palabra2):
    rango=len(palabra1)

#%% Para cambiar los caracteres especiales
    
especiales=('áéíóúüñ')
normales=('aeiouun')

for i in range (len(especiales)):
    
    for j in range (len(palabra1)):
    
         if especiales[i] == palabra1[j]:
             palabra1 = palabra1.replace(especiales[i],normales[i])
             
for i in range (len(especiales)):
    
    for j in range (len(palabra2)):
    
         if especiales[i] == palabra2[j]:
             palabra2 = palabra2.replace(palabra2[j],normales[i])
     
#%% 

for i in range (rango):
    
    if 97 <= ord(palabra1[i]) <= 122 and 97 <= ord(palabra2[i]) <= 122:
    
        if ord(palabra1[i]) < ord(palabra2[i]):
            print('La palabra ',palabra1,' está primero en el diccionario.')
            break
        
        
        elif ord(palabra1[i]) > ord(palabra2[i]):
            print('La palabra ',palabra2,' está primero en el diccionario.')
            break
        
        elif len(palabra1) < len(palabra2):
            print('La palabra ',palabra1,' está primero en el diccionario.')
            break
        
        elif len(palabra1) > len(palabra2):
            print('La palabra ',palabra2,' está primero en el diccionario.')
            break
        
        else:
            print('Las palabras son iguales')
            break
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Jun  8 19:12:42 2019

@author: SergioIdarraga
"""

# 13 Algoritmo que recibe una cadena de texto y la imprime en forma de
#árbol

cadena=input('Inserte cadena de texto: ')

# Se le quitan todos los espacios a la cadena

cad_sin_esp = cadena.replace(' ', '')

longitud = len(cad_sin_esp)

num_car_renglon = 1 # No. de caracteres de la base del triángulo.
cont_renglones = 0


while True:
    longitud -= num_car_renglon
    if longitud <= 0:
        break
    cont_renglones += 1
    num_car_renglon += 2

# En este punto, la variable longitud habrá cambiado su valor, así que se le
# vuelve a asignar el que tenía inicialmente.


longitud = len(cad_sin_esp)


num_car_renglon -= 2  # Tras el último ciclo esta variable quedó con un exce-
# dente de 2. Se compensa restándoselo.


start = 0

paso = 1


for i in range(0, cont_renglones):
    print(cad_sin_esp[start:start+paso].center(num_car_renglon))
    start += paso  # El inicio del slicing para el siguiente renglón se incre-
    # menta según la variable "paso"
    paso += 2  # Y para cada rengón se imprimen 2 caracteres más, por esto, se
    # aumenta la variable paso de 2 en 2.


# Para imprimir la última línea:


suma = 0  # Cuántos caracteres se han podido imprimir en el triángulo.
for i in range(1, num_car_renglon+1, 2):
    suma += i


if suma < longitud:  # Verifica si quedaron caracteres sobrantes por fuera del
                     # triángulo.
    print(cad_sin_esp[suma:longitud].center(num_car_renglon))
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 18 09:02:27 2019

@author: SergioIdarraga
"""

# 14 Leer una cadena de texto y organice alfabéticamente cada una de las letras
#que la componen, repitiendo cada una tantas veces como se encuentra. 
#Por ejemplo, la cadena ‘tarea importante’ será ‘aaaeeimnoprrttt’. 
#(Note que no se incluyen los espacios).


cad=input('Inserte la cadena que desea organizar alfabeticamente: ')#cadena
cad=cad.lower()#cadena ingresada en minusculas
alfabetica=[]#lista a la que se agregara el codigo unicode de cada elemento
ordenalfa=[]#lista donde se pondran en orden los elementos segun su codigo
for i in cad:
    alfabetica.append(ord(i))
ordennum=sorted(alfabetica)
for i in ordennum:
    ordenalfa.append(chr(i))#en el orden que estan, cada elemento pasa a str
    
print('La cadena ordenada alfabeticamente es: ',''.join(ordenalfa),end='')
```
